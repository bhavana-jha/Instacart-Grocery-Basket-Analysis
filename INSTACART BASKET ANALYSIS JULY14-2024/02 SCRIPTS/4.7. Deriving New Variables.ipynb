# 3. If-Statements with User-Defined Functions

```python
# Creating a subset of "ords_prods_merge" dataframe because the original one is quite large is size. 
df=df_ords_prods_merge[:1000000]
```

```python
df.head()
```

```python
df.shape
```

```python
# Defining If-Statements with User-Defined Functions
def price_label(row):

  if row['prices'] <= 5:
    return 'Low range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid range product'
  elif row['prices'] > 15:
    return 'High range product'
  else: return 'np.nan'
```

```python
# Creating a new column 'price_range' in the dataframe where the price_label output is stored. Axis=1 means "rows" and Axis=0 means 'columns'
df['price_range']=df.apply(price_label,axis=1)
```

```python
df['price_range'].value_counts(dropna=False)
```

```python
df['prices'].max()
```

# 4. If-Statements with the loc() Functions

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low range product'
```

```python
df['price_range_loc'].value_counts(dropna=False)
```

# 6. Which day most orders took place?

```python
df_ords_prods_merge['order_dow'].value_counts(dropna=False)
```

```python
# Arranging days in Busiest, Least and Regularly Categories:
result = []

for value in df_ords_prods_merge["order_dow"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# Printing result
result
```

```python
# Adding new column to df_ords_prods_merge as result
df_ords_prods_merge['busiest_day']=result
```

```python
df_ords_prods_merge['busiest_day'].value_counts(dropna=False)
```
