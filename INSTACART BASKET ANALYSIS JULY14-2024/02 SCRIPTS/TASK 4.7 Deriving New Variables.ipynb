# 4. Task Question 2

```python
# Finding value.count of each day
df_ords_prods_merge['order_dow'].value_counts(dropna=False)
```

```python
# Arranging days into three categories: Busiest Days, Slowest days and Regular Days:
result = []

for value in df_ords_prods_merge["order_dow"]:
  if value == 0 or value== 1:
    result.append("Busiest Days")
  elif value == 4 or value== 3:
    result.append("Slowest Days")
  else:
    result.append("Regular Days")
```

```python
# Printing result
result 
```

```python
# Adding new column to df_ords_prods_merge as result
df_ords_prods_merge['busiest_days']=result
```

# 6. Task Question 4 Finding the busiest hours of the day

```python
# Finding value.count of each hour
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna=False)
```

```python
# Arranging hours in three categories: "Most orders", "Average Orders", and "Fewest Orders":
result_2=[]
for value in df_ords_prods_merge["order_hour_of_day"]:
  if value == 10 or value==11 or value==14 or value==15 or value==13 or value==12 or value==16 or value==9:
    result_2.append("Most orders")
  elif value == 17 or value==8 or value==18 or value==19 or value==20 or value==7 or value==21 or value==22:
    result_2.append("Average Orders")
  else:
    result_2.append("Fewest Orders")
```

```python
# Prinitng result_2
result_2
```

```python
# Adding new column to df_ords_prods_merge as result_2
df_ords_prods_merge['busiest_period_of_day']=result_2
```
