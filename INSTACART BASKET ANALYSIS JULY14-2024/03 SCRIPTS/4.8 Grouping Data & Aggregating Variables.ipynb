# 4. Aggregating Data with agg()

```python
# Grouping Data on "department_id" and aggregating "order_number" to find the mean value
df.groupby('department_id').agg({'order_number':['mean']})
```

The list shows the average number of orders per user for each department ID

```python
# Using just mean() command to find the above result
df.groupby('department_id')['order_number'].mean()
```

The result is same as above only the output format is different. However, this does give out the data type of the value which could be more helpful than the previous command.

```python
#Performing Multiple Aggregations
df.groupby('department_id').agg({'order_number':['mean','min','max']})
```

```python
# Finding minimum mean
df.groupby('department_id')['order_number'].mean().min()
```

```python
# Finding maximum mean
df.groupby('department_id')['order_number'].mean().max()
```

# 5. Aggregating Data with Transform()

```python
# Creating Flag of "Loyal Customers"
ords_prods_merge['max_order']=ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

# 6. Deriving Columns with loc()

```python
# Creating Loyalty Flag 
ords_prods_merge.loc[ords_prods_merge['max_order']>40,'loyalty_flag']='Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order']<=40)&(ords_prods_merge['max_order']>10),'loyalty_flag']='Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order']<=10,'loyalty_flag']='New customer'
```
