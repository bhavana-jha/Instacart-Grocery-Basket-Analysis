# 3. Aggregating Data with agg()

```python
# Grouping Data on "department_id" and aggregating "order_number" to find the mean value
ords_prods_merge.groupby('department_id').agg({'order_number':['mean']})
```

```python
#Performing Multiple Aggregations
ords_prods_merge.groupby('department_id').agg({'order_number':['mean','min','max']})
```

```python
# Finding minimum mean
ords_prods_merge.groupby('department_id')['order_number'].mean().min()
```

```python
# Finding maximum mean
ords_prods_merge.groupby('department_id')['order_number'].mean().max()
```

OBSERVATIONS:
1. Minimum mean value of the entire dataframe(15.2) is more than the subset(14.3).
2. Maximum mean value of the entire dataframe(22.9) is less than the subset(25.5).
3. Each department_id has seen some level of change in the final output. Most of the department_ids' values changed in decimals, however there are few department_ids like 10(+1.5),12(+1.5), and 22(-2.6), saw noticiable changes when the command in used on the entire dataframe.
4. The minimum value remained same in both the case. The minimum order_number is 1.
5. The maximum value also remained same in both the cases except for three department_id. Those ids are 2, 8 and 21. When aggregated the entire dataset the max order_number for every department_id is 99. Howeever, in subset the max order_number for department_id 2, 8 and 21 were 96, 98 and 98 respectively.

# 7.Question 6

```python
# Finding basic info about Price column
ords_prods_merge['prices'].describe()
```

```python
# Creating "mean_product_price" column in the dataset
ords_prods_merge['mean_product_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Printing first 15 rows of the newly formed column in the dataset
ords_prods_merge[['user_id', 'mean_product_price']].head(15)
```

```python
# Creating "Spending_flag" and their value_counts
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] >= 10, 'spending_flag'] = 'High spender'
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] < 10, 'spending_flag'] = 'Low spender'
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

# 8.Question 7

```python
# Creating "median_days_between_orders" columns 
ords_prods_merge['median_days_between_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Printing first 15 rows of the newly formed column
ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days_between_orders']].head(15)
```

```python
# Creating 'order_frequency' flag
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] > 20, 'order_frequency'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_days_between_orders'] <= 20) & (ords_prods_merge['median_days_between_orders'] > 10), 'order_frequency'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] <= 10, 'order_frequency'] = 'Frequent customer'
```

```python
# Calculating value_counts of order_frequency column
ords_prods_merge['order_frequency'].value_counts(dropna = False)
```

```python
# Printing first 15 rows of the relevant columns
ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days_between_orders', 'order_frequency']].head(15)
```
